```python
from sklearn.linear_model import LinearRegression
```

```python
df = combined_df.copy()
df_yearly_total = df.groupby('year')['Total'].sum()

# ready data for linear reg
X = df_yearly_total.index.values.reshape(-1, 1)
y = df_yearly_total.values

# Fit data to model
model = LinearRegression()
model.fit(X, y)

# Pred the next years total medical staff
next_year = df_yearly_total.index.max() + 1
predicted_next_year = model.predict([[next_year]])
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_yearly_total.index, df_yearly_total, marker='o', color='blue', label='Original Data')
plt.plot([next_year], predicted_next_year, marker='o', color='red', label='Predicted Next Year')
plt.title('Trend of Total Medical Staff in Sri Lanka', color='white')
plt.xlabel('Year', color='white')
plt.ylabel('Total Medical Staff', color='white')
plt.legend(fontsize='large', loc='upper left', frameon=False, bbox_to_anchor=(1, 1), labelcolor='white')
plt.grid(True, color='gray')

plt.savefig('medical_staff_trend.png', transparent=True)

# Show the plot
plt.show()

print(f"Predicted total medical staff for the next year ({next_year}): {predicted_next_year[0]:.2f}")
```
